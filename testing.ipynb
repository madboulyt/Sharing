```python
import os
import json
from tqdm import tqdm

def load_all_json_recursive(base_dir):
    """
    Recursively scan base_dir for .json files
    and return a list of parsed JSON objects (dicts).
    """
    all_data = []

    for root, _, files in os.walk(base_dir):
        for file in tqdm(files):
            if file.lower().endswith(".json"):
                json_path = os.path.join(root, file)

                try:
                    with open(json_path, "r", encoding="utf-8") as f:
                        data = json.load(f)
                        all_data.append(data)
                except Exception as e:
                    print(f"Failed to read {json_path}: {e}")

    return all_data
```

```python
docs = load_all_json_recursive("data_10")
```

```python
# printing not done ele
for ele in docs:
    if ele["status"] != "done":
        print(ele)
```
